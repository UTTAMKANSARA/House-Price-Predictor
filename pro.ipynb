```python
def Converter(a):

    temp = a.split(' - ')
    if len(temp) == 2:
        return (float(temp[0])+float(temp[1]))/2
    try:
        return float(a)
    except:
        return None
```

```python
def removeoutliers(Y):
    df_out = pd.DataFrame()
    for key,subdf in Y.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        gen_df  =  subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft <= (m+st))]
        df_out =  pd.concat([df_out,gen_df],ignore_index = True)
    return df_out
new = removeoutliers(new)
new.describe()
```

```python
def bhkoutlier(x):
    exclude_indices = np.array([])
    for location,location_in_df in  x.groupby('location'):
        bhk_stats = {}
        for  bhk, bhk_df in location_in_df.groupby('BHK'):

            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_in_df.groupby('BHK'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count'] > 5:
                exclude_indices = np.append(exclude_indices,bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
        
    return x.drop(exclude_indices,axis ='index')
```

```python
X = new.drop(columns = ['price'])
y = new['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.metrics import r2_score
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size = 0.2,random_state =0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

## apply linear regression

```python
column_trans = make_column_transformer((OneHotEncoder(sparse_output = False),['location']),remainder ='passthrough')
```

```python
scaler = StandardScaler()
```

```python
model1 = LinearRegression()
```

```python
pipe_lr = make_pipeline(column_trans,scaler,model1)
```

```python
pipe_lr.fit(X_train,Y_train)
```

```python
y_pred = pipe_lr.predict(X_test)
```

```python
print(r2_score(Y_test,y_pred))
```
